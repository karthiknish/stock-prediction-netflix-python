```python
import pandas as pd
data = pd.read_csv('./NFLX.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
# Convert 'Date' to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Check the data types again to confirm the conversion
data.dtypes
```

```python
data.describe()
```

```python
import matplotlib.pyplot as plt

# Plot the closing price over time
plt.figure(figsize=(14,7))
plt.plot(data['Date'], data['Close'])
plt.title('Netflix Closing Stock Price Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.grid(True)
plt.show()
```

```python
# Plot the volume over time
plt.figure(figsize=(14,7))
plt.plot(data['Date'], data['Volume'])
plt.title('Netflix Trading Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.grid(True)
plt.show()
```

```python
# Plot a histogram of the closing prices
plt.figure(figsize=(10,6))
plt.hist(data['Close'], bins=30, alpha=0.75)
plt.title('Distribution of Netflix Closing Stock Prices')
plt.xlabel('Closing Price (USD)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
